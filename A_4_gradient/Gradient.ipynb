```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return (x+3)**2
```

```python
def gradient(x):
    return 2*(x+3)
```

```python
x = 2
learning_rate = 0.1
iteration = 20
```

```python
x_values = [x]
y_values = [function(x)]
```

```python
for i in range(iteration):
    grad = gradient(x)
    x = x - learning_rate*grad

    x_values.append(x)
    y_values.append(function(x))
    print(f"Iteration {i+1}: x ={x},y={function(x)}")
```

```python
x_range = np.linspace(-6,6,100)
y_range = function(x_range)
```

```python
plt.plot(x_range,y_range,label="y=(x+3)^2")
plt.scatter(x_values,y_values,color="red",label="gradient descent")
plt.plot(x_values,y_values,color='red',linestyle='--',alpha=0.6,label="path")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent to Find Local Minima")
plt.legend()  # Display the legend with proper labels
plt.show()
```
